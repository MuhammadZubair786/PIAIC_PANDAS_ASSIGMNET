# Part One Solution 

```python
# load all csv files using into a list of DataFrames using glob and list comprehension
Data_list = [pd.read_csv(csv) for csv in sorted(glob('Assignment/states[0-9].csv'))]
```

```python
#concatenate all dataframes into single dataframe
Single_Frame = pd.concat(Data_list)
```

```python
Single_Frame.head()
```

```python
#drop unwanted column and reset index
Single_Frame.drop('Unnamed: 0', axis=1, inplace=True)
Single_Frame.reset_index(drop=True, inplace=True)
```

```python
Single_Frame.head(10)
```

```python
Single_Frame.columns
```

```python
Single_Frame.dtypes
```

```python
#removing $ sign from Income column and convert to float
Single_Frame.Income = Single_Frame.Income.str.replace('$','').astype(float)
Single_Frame.dtypes
```

```python
Single_Frame.GenderPop.head()
```

```python
#splitting GenderPop column into Men and Women Columns
genderpop = Single_Frame.GenderPop.replace(['M','F'],'', regex=True).str.split('_')
men, women = genderpop.apply(lambda x: x[0]), genderpop.apply(lambda x: x[1])
Single_Frame['Men'], Single_Frame['Women'] = [men, women]
```

```python
#change newly created columns Men and Women data type to numeric
Single_Frame.Men, Single_Frame.Women = pd.to_numeric(Single_Frame.Men), pd.to_numeric(Single_Frame.Women)
Single_Frame[['Men', 'Women']]
```

```python
# now dealing with NaN values in women population by filling it with values (TotalPop - Men)
Single_Frame.Women.fillna(Single_Frame.TotalPop - Single_Frame.Men, inplace=True)
Single_Frame[['Men', 'Women']]
```

```python
#plotting a scatter graph of Women and Income
plt.scatter(Single_Frame.Women, Single_Frame.Income, color=['orange'])
plt.xlabel('Women')
plt.ylabel('Income')
plt.show()
```

```python
# check duplicated columns
Single_Frame.duplicated() .value_counts()
```

```python
# drop duplicated rows and reset index to see the difference
Single_Frame.drop_duplicates(inplace=True)
Single_Frame.reset_index(drop=True, inplace=True)
Single_Frame.duplicated() .value_counts()
```

```python
#plot again to see a perfect plot
plt.scatter(Single_Frame.Women, Single_Frame.Income, color=['red'])
plt.xlabel('Women')
plt.ylabel('Income')
plt.show()
```

```python
# races columns
Single_Frame.iloc[:, 2:8].head()
```

```python
Single_Frame.iloc[:, 2:8] = Single_Frame.iloc[:, 2:8].replace('%','', regex=True).astype(float)
```

```python
#find nan values in races which are in only in Pacific race
Single_Frame.Pacific.loc[Single_Frame.Pacific.isnull() == True]
```

```python
#fill nan with mean of Pacific race
Single_Frame.Pacific.fillna(Single_Frame.Pacific.mean(), inplace=True)
Single_Frame.Pacific.loc[Single_Frame.Pacific.isnull() == True]
```

```python
# function to create Histograms of various races of US
def RaceHistogram(race, color):
    plt.hist(Single_Frame[race],ec='black',color=color)
    plt.title(race+" Race")
    plt.ylabel("Frequency")
    plt.xlabel("Percentage of Population")
    plt.show()
```

```python
# Hispanic
RaceHistogram('Hispanic', 'chocolate')
# White
RaceHistogram('White', 'whitesmoke')
# Black
RaceHistogram('Black', 'dimgrey')
# Native
RaceHistogram('Native', 'coral')
# Asian
RaceHistogram('Asian', 'saddlebrown')
# Pacific
RaceHistogram('Pacific', 'tan')
```

```python
# read inventory.csv
inventory = pd.read_csv('Assignment/inventory.csv')
inventory.head(10)
```

```python
# Brooklyn and seeds
seed_request = inventory.loc[(inventory['location'] == 'Brooklyn') & ( inventory['product_type'] == 'seeds')]
seed_request
```

```python
# create new in_stock column
inventory['in_stock'] = inventory.quantity != 0
inventory.head(10)
```

```python
# create new col total value
inventory['total_value'] = inventory.price * inventory.quantity
inventory.head(10)
```

```python
#full description column
combine_lambda = lambda row:'{} - {}'.format(row.product_type, row.product_description)
inventory['full_description'] = inventory.apply(combine_lambda, axis=1)
inventory
```
